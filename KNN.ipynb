```python
import numpy as np
import cv2
import os
from imutils import paths
```

```python
imagepaths = list(paths.list_images("/Users/tharun/Desktop/ML/Garbage Dataset/train"))
```

```python
imagepaths
```

```python
data_knn = []
labels_knn = []
for images in imagepaths:
    label = images.split(os.path.sep)[-2]
    image =cv2.imread(images)
    image =cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    image =cv2.resize(image, (200,200))
    image = np.asarray(image)
    image = np.reshape(image,(120000))
    data_knn.append(image)
    labels_knn.append(label)
```

```python
print(len(data_knn))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(data_knn,labels_knn)
```

```python
testpaths = list(paths.list_images("/Users/tharun/Desktop/ML/Garbage Dataset/test"))
```

```python
test_data_knn = []
test_labels_knn = []
for images in testpaths:
    label = images.split(os.path.sep)[-2]
    image =cv2.imread(images)
    image =cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    image =cv2.resize(image, (200,200))
    image = np.asarray(image)
    image = np.reshape(image,(120000))
    test_data_knn.append(image)
    test_labels_knn.append(label)
```

```python
from sklearn.metrics import accuracy_score
pred_labels = model.predict(test_data_knn)
print(pred_labels)
```

```python
accuracy_score(pred_labels,test_labels_knn)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(pred_labels,test_labels_knn)
```
